```python
import os
import json
import random

base_folder = '../../metric_validation_0'

# Sample a random set of 100 tables to get alignment annotations for
file_list = [x for x in os.listdir(base_folder) if x.endswith('json')]
chosen_ids = random.sample(file_list, 100)
```

```python
# Store chosen IDs in output file for record
out_file = open('alignment_files.txt', 'w')
for id in chosen_ids:
    out_file.write(id.split('_')[0]+'\n')
out_file.close()
```

```python
reader = open('alignment_files.txt')
chosen_ids = []
for row in reader:
    chosen_ids.append(row.strip())
print(chosen_ids)
```

```python
# Get gold table and all prediction tables for each ID
import pandas as pd
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
out_folder = '../annotations/goldcaption_column_alignment'

counter = 0
for id in chosen_ids:
    gold_table_file = json.loads(open(os.path.join(base_folder, id)).read())
    gold_table_df = pd.DataFrame(gold_table_file["table"])
    gold_table_schema = list(gold_table_file["table"].keys())
    id_val = id.split('_')[0]
    pred_paths = [
        os.path.join(base_folder, id_val, 'gpt3.5/baseline_outputs/try_0.json'),
        os.path.join(base_folder, id_val, 'gpt3.5/ours_outputs/try_0.json'),
        os.path.join(base_folder, id_val, 'mixtral/baseline_outputs/try_0.json'),
        os.path.join(base_folder, id_val, 'mixtral/ours_outputs/try_0.json')
    ]
    for path in pred_paths:
        table_data = json.loads(open(path).read())
        for table in table_data:
            table_df = pd.DataFrame(table["table"])
            table_schema = table["schema"]
            print(f"Generating file num {counter}")
            out_file = open(os.path.join(out_folder, f'{counter}_alignment.txt'), 'w')
            out_file.write(f"ID: {id_val}\n")
            out_file.write("\nGOLD TABLE:\n")
            out_file.write(gold_table_df.to_markdown())
            out_file.write("\n\nGOLD SCHEMA:\n")
            out_file.write("\n".join([f"{x}: {y}" for x,y in zip(list(range(0, len(gold_table_schema))), gold_table_schema)]))
            out_file.write(f"\n\nPREDICTION PATH:{path}")
            out_file.write("\n\nPREDICTED TABLE:\n")
            out_file.write(table_df.to_markdown())
            out_file.write("\n\nMATCHES:\n")
            out_file.write("\n".join([x+": " for x in table_schema]))
            out_file.close()
            counter += 1
```
